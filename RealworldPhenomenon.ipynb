##### To have a clear picture of the data points I will use I did a research on variables. 

### Shoe Size
I decided to use European women’s shoe size which goes from size 35 to size 42. For reference please see website which describes [SIZE GUIDE]( https://m.asos.com/Women/footwear-size-guide/?szgid=8&szgtid=1&r=2).
### Shoe Type
I have done some research on few website and I came to conclusion the shoe types which are commonly on sales are Trainers, High Heels, Boots and Sandals. I took this information from visiting Sale section of different websites like [Office](https://www.office.co.uk/sale). As we are in December, [research](http://brokeandchic.com/when-is-the-best-time-to-buy-boot/) shows that best time to buy boots is from September to December. 
### Original Prize
I have decided to set a price range that is affordable. As all my previous research was based on middle class classification I decided to take a price range from a shoe website [Topshop](http://eu.topshop.com/en/tseu/category/shoes-485095).  The prize will vary from 5euro to 195euro.
### Discount of 20%
I have decided that since we are in December months the discount on average has to match the year sale cycle which is usually higher at the end of the year. [MoneyGuideIreland](http://www.moneyguideireland.com/sales-starting-december-26th.html) wrote that Schuh shoe shop start their sale in December with up to 45 per cent off styles from the shoe shop’s autumn-winter collections. As my dataset is not exclusively autumn-winter collection but more a variety of collections I have decided to assign a discount of 20%.
### Sale Prize
This will give me the sale prize of the shoe after 20% discount is taken away from the original prize
### Amount Available
Lastly, this data will give us an idea of the amount of shoe pairs in a particular size and shoe type there is in the shop.

```python
#Im going to focus on fist column first 'shoe size'
#As this is women size it starts from 35 to 42
#this will allocate the sizes at random
data = pd.DataFrame({"shoe_size": np.random.randint(35,43,100)}) 
#this will generate 100 random numbers between smallest size of shoe 35 and largest 42.
data.tail(10) #shows last 10 data points and to show the first 10 data we can use data.head(10)
```

```python
#Now I want to focus on the shoe type, this code will have to generate four different shoe types
#the shoe types I choosed are trainers, sandals, high heels and boots.
types = ['trainers', 'sandals', 'boots', 'high_heels']

data = pd.DataFrame({"shoe_size": np.random.randint(35,43,100), "shoe_type": np.random.shuffle(types)})
data.head(10)
#I received an error on shoe_type
#I believe it is because the amount of times has to be identified and shuffle is not a correct function 
#random.shuffle
```

```python
#random.choice
data = pd.DataFrame({"shoe_size": np.random.randint(35,43,100), "shoe_type": np.random.choice(types, 100)})
data.head(20)
#I used random choice as it allows multiple elements from a sequence with replacement (duplicates are possible)
```

As I am concentrating on middle class range of shoes I checked a website called [Topshop](http://eu.topshop.com/en/tseu/category/shoes-485095/N-a2eZdgf?No=0&Nrpp=20&siteId=%2F13058&Nf=nowPrice%7CBTWN%2B5%2B195) which sells shoes and the prices vary from 5 euro to 195 euro. That is the range I will allocate for my shoes before the discount.

```python
#randomly allocating a range of numbers between 5 euro and 195 euro.
dataprice = pd.DataFrame({"original_price": np.random.randint(5,195,100)}) 
dataprice.head(15)
```

```python
#now lets put it into the draft dataset:
data = pd.DataFrame({"shoe_size": np.random.randint(35,43,100), 
                     "shoe_type": np.random.choice(types, 100), "original_price": np.random.randint(5,195,100)})
data.head()
```

```python
#this column will show what is 20% dicount from the orginal prize
data['discount_20'] = data['original_price'] * 20 // 100 
data
data.head()
```

```python
#this additional column will show the sale price after 20% discount is applied on the original price
#Here I researched how to subtract two columns in a dataframe
data['sale_price'] = data['original_price']-data['discount_20']
data.head()
```

```python
#I wanted to have a link between the type of shoe type and amount available by mapping two columns:
#data['setamount_available'] = data['shoe_type'].map({'sandals': 3, 'trainers': 10, 'boots': 20, 'high_heels': 5})
#data.loc[0:10, ['shoe_type', 'setamount_available']]
#I did not end up using mapping as it turned out later it is not the best method when analysing the data later
```

```python
#I have then decided to randomly assign number between 1 to 10 pairs available at random.
data['amount_available'] = np.random.randint(1,10,100) #random amount from 1 to 10 pairs left on sale?
data.head()
```

```python
#Now this is the output of the overall data
data
```
